```python
import pandas as pd

# Define the chunk size
chunk_size = 10000

# Initialize dictionaries to hold missing values count and unique values for each column
missing_values_count = {}
unique_values = {}

# Initialize variables for tracking the total number of rows and columns
total_rows = 0
total_columns = None  # Will set it once the first chunk is read

# Iterate through each chunk once for both operations
for chunk_num, chunk in enumerate(pd.read_csv('Training_final/Additional Features.csv', chunksize=chunk_size)):
# for chunk_num, chunk in enumerate(pd.read_csv('/Users/shreya/FALL24/TAMU/Processing/EDA/processed_dataset_add_features.csv', chunksize=chunk_size)):
    # Print the shape of each chunk
    # print(f"Processing chunk {chunk_num + 1} with shape: {chunk.shape}")

    # Update the total rows
    total_rows += chunk.shape[0]

    # Set the total_columns once for the first chunk
    if total_columns is None:
        total_columns = chunk.shape[1]

    # Count missing values
    chunk_missing = chunk.isnull().sum()
    for column, count in chunk_missing.items():
        if column in missing_values_count:
            missing_values_count[column] += count
        else:
            missing_values_count[column] = count

    # Count unique values
    for column in chunk.columns:
        if column in unique_values:
            unique_values[column].update(chunk[column].dropna().unique())
        else:
            unique_values[column] = set(chunk[column].dropna().unique())

# Calculate the number of unique values for each column
unique_counts = {col: len(val_set) for col, val_set in unique_values.items()}
```

```python
import pandas as pd
from sklearn.impute import KNNImputer, SimpleImputer

# Define the chunk size
chunk_size = 10000

# Initialize imputers
knn_imputer = KNNImputer(n_neighbors=5)
median_imputer = SimpleImputer(strategy='median')
mode_imputer = SimpleImputer(strategy='most_frequent')

# Specify columns for each imputation method
knn_columns = ['cci_score', 'dcsi_score', 'fci_score', 'cms_tot_partd_payment_amt',
               'cms_tot_ma_payment_amt', 'atlas_grocpth14', 'atlas_ffrpth14', 'atlas_fsrpth14']
median_columns = ['atlas_povertyallagespct', 'atlas_recfacpth14']
mode_columns = ['cms_frailty_ind']

# Process the dataset in chunks
for chunk in pd.read_csv('/content/Training_final/Additional Features.csv', chunksize=chunk_size):

    # Impute KNN columns in the chunk
    chunk[knn_columns] = knn_imputer.fit_transform(chunk[knn_columns])

    # Impute median columns
    chunk[median_columns] = median_imputer.fit_transform(chunk[median_columns])

    # Impute mode columns
    chunk[mode_columns] = mode_imputer.fit_transform(chunk[mode_columns])

    # Save or append the processed chunk
    chunk.to_csv('processed_add_features', mode='a', header=False, index=False)
```

## Remvoing "YEARS" from the tenure_band column and taking avg of value (removing hyphen)

```python
import pandas as pd

# Define the chunk size
chunk_size = 10000

# Function to process tenure_band column
def clean_tenure_band(tenure_band):
    # Remove 'YEARS' and strip whitespace
    tenure_band = tenure_band.replace("YEARS", "").strip()

    # Split by hyphen and calculate average
    try:
        start, end = map(float, tenure_band.split('-'))
        return (start + end) / 2
    except ValueError:
        return None  # Handle any invalid formats

# Process each chunk
for chunk_num, chunk in enumerate(pd.read_csv('Training_final/MEMBER_DATA.csv', chunksize=chunk_size)):
    # Apply the cleaning function to the tenure_band column
    chunk['tenure_band'] = chunk['tenure_band'].apply(clean_tenure_band)

    # Drop rows where tenure_band could not be processed
    chunk = chunk.dropna(subset=['tenure_band'])

    # Save the cleaned chunk to a new file
    chunk.to_csv('processed_MEMBER_DATA.csv', mode='a', header=(chunk_num == 0), index=False)
```
